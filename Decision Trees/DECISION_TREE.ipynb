```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("salary_data1.csv")
data.keys()
data = data.replace(r'\\n','', regex = True)
data
```

```python
data["Prev Designation"] = data["Prev Designation"].str.strip()
data["College"] = data["College"].str.strip(r'\\n')
```

```python
data["Prev Designation"] = data["Prev Designation"].str.lower()
```

```python
data["Designation"] = "current" + data["Designation"]
data["Prev Designation"] = "prev" + data["Prev Designation"]
data["Prev Designation"].value_counts()
```

```python
temp = pd.get_dummies(data["College"])
data = data.drop("College", axis = 1)
data = data.join(temp)
data
```

```python
temp = pd.get_dummies(data["Prev Designation"])
data = data.drop("Prev Designation", axis=1)
data = data.join(temp)
data
```

```python
temp = pd.get_dummies(data["Degree"])
data = data.drop("Degree", axis=1)
data = data.join(temp)
data
```

```python
temp = pd.get_dummies(data["Industry Domain"])
data = data.drop("Industry Domain", axis=1)
data = data.join(temp)
data
```

```python
temp = pd.get_dummies(data["Designation"])
data = data.drop("Designation", axis=1)
data = data.join(temp)
data
```

```python
from sklearn.model_selection import train_test_split 
y = data["Salary(PM)"]
x = data.drop("Salary(PM)", axis = 1)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,shuffle = False)
print(len(x_train))
print(len(x_test))
print(len(y_train))
print(len(y_test))
#print(x_test)
```

```python
from sklearn.tree import DecisionTreeRegressor as DTR
from sklearn.metrics import mean_squared_error
import numpy as np

rmse_list=[]
for i in range(1,13):
    dt = DTR(max_depth=i)
    dt.fit(x_train,y_train)
    y_pred = dt.predict(x_test)
    rmse_list.append(np.sqrt(mean_squared_error(y_test, y_pred)))
    
print(rmse_list)
```

```python
min(rmse_list)
```

```python
max(rmse_list)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(range(1,13),rmse_list)
```
