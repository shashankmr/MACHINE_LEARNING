```python
import pandas as pd
```

```python
data = pd.read_csv("salary_data1.csv")
data.keys()
data = data.replace(r'\\n','', regex = True)
data
```

```python
data["Prev Designation"] = data["Prev Designation"].str.strip()
data["College"] = data["College"].str.strip(r'\\n')
```

```python
data["Prev Designation"] = data["Prev Designation"].str.lower()
```

```python
data["Designation"] = "current" + data["Designation"]
data["Prev Designation"] = "prev" + data["Prev Designation"]
data["Prev Designation"].value_counts()
```

```python
data["Designation"].value_counts()
data["College"].value_counts()
```

```python
temp = pd.get_dummies(data["College"])
data = data.drop("College", axis = 1)
data = data.join(temp)
```

```python
temp = pd.get_dummies(data["Prev Designation"])
data = data.drop("Prev Designation", axis=1)
data = data.join(temp)
```

```python
temp = pd.get_dummies(data["Degree"])
data = data.drop("Degree", axis=1)
data = data.join(temp)
```

```python
temp = pd.get_dummies(data["Industry Domain"])
data = data.drop("Industry Domain", axis=1)
data = data.join(temp)
```

```python
temp = pd.get_dummies(data["Designation"])
data = data.drop("Designation", axis=1)
data = data.join(temp)
data
```

```python
from sklearn.model_selection import train_test_split 
y = data["Salary(PM)"]
x = data.drop("Salary(PM)", axis = 1)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,shuffle = False)
print(len(x_train))
print(len(x_test))
print(len(y_train))
print(len(y_test))
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt_regressor = DecisionTreeRegressor(max_depth = 17)
dt_regressor.fit(x_train,y_train)
```

```python
y_predict = dt_regressor.predict(x_test)
y_predict
```

```python
y_test
```

```python
dt_regressor.score(x_train,y_train)
```

```python
dt_regressor.score(x_test,y_test)
```

```python
from sklearn import tree
from IPython.display import Image
import pydotplus
dot_data = tree.export_graphviz(dt_regressor, out_file = None)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
graph.write_png("salary_data14.png")
```
